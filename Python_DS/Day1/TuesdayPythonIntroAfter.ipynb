# Functions

```python
def addition(a,b):
    print(a)
    print(b)
    return a+b

c = addition([2,3],[5,6])
print(c)
```

Functions can return multiple values - bundled into a tuple.

```python
def swapFunction(a,b):
    return b,a, "Martina"

result = swapFunction(3,4)
print(type(result))
a,b,c = result
print(a, b, c)
```

Function parameters can have default values and can be named

```python
from math import sqrt

def quadratic(a, b, c):
    x1 = -b / (2*a)
    x2 = sqrt(b**2 - 4*a*c) / (2*a)
    return (x1 + x2), (x1 - x2)

#parameters passed positionally
quadratic(1,7,3)

quadratic (a=1, c=7, b=19)
```

```python
#lambda functions = anonymous functions
addition = lambda a,b: a+b

double = lambda x: x*2

res = addition(4,6)
print(res)
```

```python
res = double(6)
print(res)
```
